```python
import pandas as pd
```

```python
dataPath = r'../data/wind_dataset.csv'
```

```python
df = pd.read_csv(dataPath, sep=',', header=0, index_col=0)
```

```python
df
```

```python
# replace 0's with NaN
df = df.replace(0, pd.NA)

# Convert all columns to numeric, coercing errors to NaN
df = df.apply(pd.to_numeric, errors='coerce')

# fill NaN values with the exponential moving average
df = df.fillna(df.ewm(span=5).mean())
```

```python
# use backward fill to fill NaN values
df = df.fillna(method='bfill')
```

```python
df
```

```python
# Rename WIND to Out
df = df.rename(columns={'WIND': 'Out'})

df.index.name = 'Date'
```

```python
df
```

```python
df.to_csv(dataPath+'_out', sep=',', header=True, index=True)
```
